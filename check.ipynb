```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Professors_extracted_data.csv")
```

```python
val = df['Designation'].value_counts()
height=[]
for i in val:
    height.append(i)
print(height)
```

```python
plt.bar(df['Designation'].unique(),height=height)
plt.xticks(rotation=90)
```
